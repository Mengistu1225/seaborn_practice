```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# sns.boxplot(data,x/y,hue,orient,width,palette,showfilters)
data=sns.load_dataset('titanic')

data
```

```python
# Create a box plot to visualize the distribution of fares
sns.boxplot(x=data['fare'])

plt.title("Box Plot of Titanic Fares")
plt.show()
```

```python
# Create a grouped box plot of fares based on class and survival status
sns.boxplot(x='class',y='fare',hue='survived',data=data,palette='Set2')
plt.title("Box Plot of Fares by Class and Survival")
plt.xlabel("class")
plt.ylabel("fare")
plt.show()
```

```python
sns.boxplot(x='class',y='fare',hue='survived',data=data,palette='Set2')
plt.title("Box Plot of Fares by Class and Survival")
plt.xlabel("class")
plt.ylabel("fare")
plt.legend(title="Survived", loc="upper right")
plt.show()
```

```python
sns.boxplot(x=data["fare"], showfliers=False)
plt.title("Box Plot of Titanic Fares (Without Outliers)")
plt.show()
```

```python
sns.boxplot(x=data["fare"], showfliers=True)
plt.title("Box Plot of Titanic Fares (Without Outliers)")
plt.show()
```

```python
# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Box plot of age grouped by class and gender
sns.boxplot(x="class", y="age", hue="sex", data=titanic, palette="coolwarm")

plt.title("Box Plot of Age by Class and Gender")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.legend(title="Gender", loc="upper right")
plt.show()
```

```python
# Horizontal box plot of age grouped by class
sns.boxplot(y="class", x="age", data=titanic, orient="h", palette="Set2")

plt.title("Horizontal Box Plot of Age by Class")
plt.xlabel("Age")
plt.ylabel("Passenger Class")
plt.show()
```

```python
# Combine box plot with swarm plot
sns.boxplot(x="class", y="age", data=titanic, palette="pastel")
sns.swarmplot(x="class", y="age", data=titanic, color="0.25", size=3)

plt.title("Box Plot with Swarm Plot of Age by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

```python
# Box plot without outliers
sns.boxplot(x="class", y="fare", data=titanic, showfliers=False, palette="Blues")

plt.title("Box Plot of Fare by Class (Without Outliers)")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

```python
# Box plot of fare
sns.boxplot(x=titanic["fare"], color="orange")

plt.title("Box Plot of Titanic Fares")
plt.xlabel("Fare")
plt.show()
```

```python


# Custom dataset
data = {
    "Category": ["A", "A", "A", "B", "B", "B", "C", "C", "C"],
    "Subcategory": ["X", "Y", "Z", "X", "Y", "Z", "X", "Y", "Z"],
    "Values": [10, 15, 14, 20, 22, 19, 30, 35, 33]
}
df = pd.DataFrame(data)

# Grouped box plot
sns.boxplot(x="Category", y="Values", hue="Subcategory", data=df, palette="coolwarm")

plt.title("Grouped Box Plot")
plt.xlabel("Category")
plt.ylabel("Values")
plt.legend(title="Subcategory")
plt.show()
```

```python
# Custom color palette
sns.boxplot(x="class", y="fare", data=titanic, palette={"First": "red", "Second": "blue", "Third": "green"})

plt.title("Box Plot of Fare by Class with Custom Colors")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

```python
# Filter Titanic data for adults only (age >= 18)
filtered_data = titanic[titanic["age"] >= 18]

# Box plot of fare for adults
sns.boxplot(x="class", y="fare", data=filtered_data, palette="cubehelix")

plt.title("Box Plot of Fare by Class (Adults Only)")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

```python
# Identify outliers in the "age" column
sns.boxplot(x=titanic["age"], color="purple")

plt.title("Outlier Detection in Age")
plt.xlabel("Age")
plt.show()
```

```python
# Box plots for age by class, split by gender
g = sns.catplot(x="class", y="age", col="sex", data=titanic, kind="box", palette="viridis")

g.fig.suptitle("Box Plot of Age by Class and Gender", y=1.02)
g.set_axis_labels("Passenger Class", "Age")
plt.show()
```
