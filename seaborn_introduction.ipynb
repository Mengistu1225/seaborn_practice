```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cars=sn.load_dataset('mpg')
```

```python
cars.head()
```

```python
cars.dropna(inplace=True)
```

```python
cars
```

```python
cars.columns
```

```python
sn.relplot(x='model_year',y='mpg',data=cars)
```

```python
sn.relplot(x='model_year',y='mpg',col='origin',data=cars)
```

```python
sn.relplot(x='model_year',y='mpg',col='origin',hue='cylinders',data=cars)
```

```python
sn.pairplot(cars[cars.cylinders.isin([4,6,8])],
           hue='cylinders',
           vars=['mpg','horsepower','acceleration'],
           palette='inferno_r');
```

```python
usa=cars[cars.origin=='usa']
sn.violinplot(x=usa.cylinders,y=usa.horsepower,color='xkcd:forest green',scale='width')
```

```python
sn.violinplot(x=usa.cylinders,y=usa.horsepower,color='xkcd:forest green',scale='width',inner=None)
sn.swarmplot(x=usa.cylinders,y=usa.horsepower,size=3,color='white')
```

```python
sn.set_style('white')
pts = [1, 1.5, (7/3), 3, 3.75, 4.25, 5, 19/3, 8, 12]

plt.figure(figsize=(6,4))
sn.boxplot(pts, color='xkcd:golden yellow', width=0.4)
sn.despine(left=True)
plt.xlim(-3, 16)
plt.xticks(range(0,13,2))
plt.tight_layout();
```

```python
plt.figure(figsize=(8,5))
sn.distplot(cars.weight, 
             bins=20,
             kde_kws={"lw": 8, 'color':'xkcd:bluish green'},
             hist_kws={'alpha': 0.3, 'color':'gray'}
            )

#sn.despine(left=True)
plt.xlabel('Weight (lbs)', fontsize=16, labelpad=15)
plt.yticks([])
plt.tight_layout();
```
